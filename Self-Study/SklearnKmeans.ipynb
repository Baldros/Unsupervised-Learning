**Nota**

KNN != KMeans:

    Cuidado para não confundir o KNN
    com o KMeans! Apesar de ambos
    trabalharem com agrupamento, são
    algorítmos que funcionam de forma
    diferente e utilizados em situações
    diferentes!


    1. Objetivo Principal:
    KNN (K-Nearest Neighbors): É um
    algoritmo de classificação usado para
    fazer previsões com base na maioria
    dos "vizinhos" mais próximos de um
    ponto de dados desconhecido.

    KMeans: É um algoritmo de agrupamento
    que separa um conjunto de dados em
    clusters (grupos) com base nas
    características dos dados, buscando
    minimizar a variação intra-cluster.


    2. Tipo de problema a ser resolvido:
    KNN: É um algoritmo supervisionado,
    pois requer rótulos para treinar o
    modelo e fazer previsões.

    KMeans: É um algoritmo não
    supervisionado, pois não requer
    rótulos e agrupa os dados com base
    em similaridades intrínsecas.
    
    3. Número de Clusters (K):
    KNN: Não envolve clusters. O valor
    de K no KNN refere-se ao número de   
    vizinhos mais próximos considerados
    durante a classificação.

    KMeans: Requer que o número de
    clusters (K) seja especificado
    antes da execução do algoritmo.

    4. Saída:
    KNN: Fornece uma previsão de classe
    para um ponto de dados desconhecido.

    KMeans: Gera clusters, cada um   
    representado por um centróide,
    que é a média dos pontos no
    cluster.

    5. Aprendizado:
    KNN: Armazena todos os dados de
    treinamento e realiza cálculos de
    distância durante a previsão.

    KMeans: Iterativamente atualiza os
    centróides dos clusters até convergir
    para uma solução.

# Preparando conjunto de dados:

    Vamos trabalhar com o mesmo
    dataset que já trabalhamos algumas
    vezes.

```python
from sklearn import datasets
```

```python
# Checando os Datasets:
for name in dir(datasets):
  if "load" in name:
    print(name)
```

```python
iris = datasets.load_iris();iris.keys()
```

```python
samples = iris.data
original_labels = iris.target
```

# K-means:

    O algoritmo KMeans agrupa dados
    tentando separar amostras em n grupos
    de variância igual, minimizando um
    critério conhecido como inércia ou
    soma dos quadrados dentro do cluster.

    Ele dimensiona bem para grandes
    quantidades de amostras e tem sido
    utilizado em uma ampla variedade de
    áreas de aplicação em diversos campos.

sphx_glr_plot_kmeans_assumptions_thumb.png


    A ideia do k-means é dividir um
    conjunto de N amostras X em K clusters
    disjuntos C, cada um descrito pela média
    das amostras no cluster. As médias são
    comumente chamadas de "centroides" do
    cluster; observe que, em geral, elas
    não são pontos das amostras, embora
    estejam no mesmo espaço.

    O algoritmo K-means visa escolher
    centróides que minimizem a inércia, ou
    critério de soma de quadrados dentro
    do cluster:

minimizem.png




```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
model.fit(samples)
```

```python
# labels:
labels = model.predict(samples);labels
```

```python
print(original_labels,original_labels.shape)
```

```python
print(labels,labels.shape)
```

```python
# Calculando a acurácia:
sum(original_labels == labels)/len(original_labels)
```

```python
# Atribuindo os centros do cluster: centróides
centroids = model.cluster_centers_;print(centroids,centroids.shape)
```

# Visualização:

    Aqui vamos elaborar algumas visualizações
    para nos ajudar no entendimento do modelo.

```python
import matplotlib.pyplot as plt
```

```python
# Separando a amostra da sépala(1º coluna de indice 0) da pétala(3º coluna de indice 2)
xs = samples[:,0]
ys = samples[:,2]
```

```python
plt.scatter(xs,ys,c=labels)
plt.show()
```

```python
centroides = model.cluster_centers_;print(centroides,centroides.shape)
```

```python
# Separando os centroides:
centroids_x = centroids[:,0]
centroids_y = centroids[:,2]
```

```python
# Visualização:
plt.scatter(xs,ys,c=labels)
plt.scatter(centroids_x,centroids_y,marker="D",s=50)
plt.show()
```
