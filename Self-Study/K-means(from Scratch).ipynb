# K-means from Scratch:

## Algoritmo:
1. Retirar escala dos dados;
  * O método escolhido para retirar estala dos dados foi a padronização:
  $z_n = \frac{x-\mu}{\sigma}$
  
2. Inicializar centroids;
3. Etiquetar cada ponto;
4. Atualizar centroids;
5. Repeti-se os passos 3 e 4 os centroids pararem de mudar.

## Retidando a estala dos dados:

```python
# Dados:
data = digits.data;data.T
```

```python
# Normalização dos Dados:
dados_normalizados = []
for data_point in data.T:
  mean = np.mean(data_point)
  std = np.std(data_point)
  data_point_normalizado = [(valor - mean)/std for valor in data_point]  # Maior precisão dos valores
  #data_point_normalizado = (data_point - mean) / std # Menor precisão dos valores
  dados_normalizados.append(data_point_normalizado)
```

```python
# Filtrando Valores Faltantes:
dadosLimpos = [sublista for sublista in dados_normalizados if not np.any(np.isnan(sublista))]
```

```python
def shape(vetor):
  return len(vetor), len(vetor[0])

print(shape(dados_normalizados), shape(dadosLimpos))
```

```python
dataTransfor = np.matrix(dadosLimpos);dataTransfor
```

```python
def initialize_centroids(X, k):
  '''
    Função construida para gerar os centroids
  de forma aleatória.

  # Entradas:
  numpy.object: Matrix dos dados.
  int: Numéro de centróides.

  # Saída:
  numpy.object: Matrix dos centróides.
  '''

  # Seleciona aleatoriamente k índices dos dados
  indices = np.random.choice(X.shape[0], size=k, replace=False)

  # Retorna os k pontos correspondentes como centróides iniciais
  centroids = X[indices]

  return centroids
```

# Etiquetando cada centroid:

```python
def distancia(matriz, mean):
  '''
    Função que calcula a distância euclidiana
  entre os pontos.

  # Entradas:
  numpy.object: Matrix dos dados.
  float: média dos dados.

  # Saída:
  numpy.object: Matrix das distâncias.
  '''

  distancia = np.sqrt(np.sum((matriz - mean)**2, axis=1))
  return distancia
```
